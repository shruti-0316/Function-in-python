**2.A python program to find prime no upto n using functions.**

---



```python
def prime_no(num):
    if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num,end=" ")
    
limit  = int(input("enter starting limit: "))

for i in range(limit):
   prime_no(i)
```

**4.A function that returns addition,substraction,multiplaction and division.**

---



```python
n1=int(input("ENter first number:"))
n2=int(input("ENter second number:"))

def add(num1,num2):
  add1=num1+num2
  print("Addition of two numbers:",add1)

def sub(num1,num2):
  sub1=num1-num2
  print("Substraction of two numbers:",sub1)


def mul(num1,num2):
  mul1=num1*num2
  print("Multiplaction of two numbers:",mul1)


def div(num1,num2):
  div1=num1/num2
  print("Division of two numbers:",div1)

add(n1,n2)
sub(n1,n2)
mul(n1,n2)
div(n1,n2)
```
